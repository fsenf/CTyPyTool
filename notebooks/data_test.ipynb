```python
#from netCDF4 import Dataset 
import numpy as np
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy
import xarray as xr


data_directory = "../data/"
#filename = "msevi_georef.nc"
filename_labels = "nwcsaf_msevi-nawdex-20160930.nc"
filename_data = "msevi-nawdex-20160925.nc"
mask_filename = ""


#sat_data = Dataset(data_directory + filename_data)

sat_data = xr.open_dataset(data_directory+filename_data)
#print(sat_data)
print(sat_data.variables)

lons = sat_data.variables['lon']

lats = sat_data.variables['lat']

for variable in sat_data.variables:
    print(variable)
bt120 = sat_data.variables['zen']




def plotData():

	central_lon = lons.mean()
	central_lat = lats.mean()
	extent = [lons.min(), lons.max(), lats.min(), lats.max()]
	ax = plt.axes(projection=ccrs.PlateCarree())
	ax.set_extent(extent, crs=ccrs.PlateCarree())
	ax.gridlines()
	ax.coastlines(resolution='50m')
	ax.add_feature(cartopy.feature.OCEAN)
	ax.add_feature(cartopy.feature.LAND, edgecolor='black')
	ax.add_feature(cartopy.feature.LAKES, edgecolor='black')
	ax.add_feature(cartopy.feature.RIVERS)	
	ax.contourf(lons, lats, bt120[0], cmap = "jet")
	plt.show()

plotData()
```
